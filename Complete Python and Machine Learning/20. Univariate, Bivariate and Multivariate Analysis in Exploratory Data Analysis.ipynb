```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df.head()
```

```python
df.shape
```

### Univariate Analysis


```python
df_setosa = df.loc[df['species']=='setosa']
```

```python
df_virginica = df.loc[df['species']=='virginica']
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'r+')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']), 'b*')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'yo')
plt.xlabel('Petal Length')
plt.show
```

## Bivariate Analysis

```python
sns.FacetGrid(df,hue="species",height=5).map(plt.scatter,"sepal_length","sepal_width").add_legend();
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df,hue="species",height=3)
```
