```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")  #load_dataset is inbuilt seaborn function
```

```python
#Here tip is actually your dependent feature and total_bill,smoker,sex,day are independent features in exploratory data analysis.
```

# Correlation with Heatmap

A correlation heatmap uses coloured cells, typically in a monochromatic scale., to show a 2D correlation matrix (table)
between two discrete dimension or event types. It is very important in Feature Selection.
It is called as Pearson's Coefficient Correlation.

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

# JoinPlot
A joinplot allows to study the relationship between 2 numeric variables. The central chart display their correlation.
It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot.

Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') #in kind=reg the graph will draw a probability Density Function to the regression line
```

# Pair Plot
A 'pairs plot' is also known as scatterplot, in which one variable in the same data row is matched with 
another variable's value, like this: Pairs plots are just ellaborations on this, showing all variables 
paired with all other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex') # hue=sex is a graph which is created by using sex(Male,Female) column to match dependent variables.
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

# Dist plot
It helps us to check the distribution of columns feature.

```python
sns.distplot(df['tip'],color='r')#By default kde is true so count is in percentage like 0.1 %,0.2% .. and so on
```

```python
sns.distplot(df['tip'],kde=False,bins=10,color='r')#kde is kernel density estimation used for PDF line
```

```python
#Notice the change in Y-axis of distplot this is due to kde being set to true or false.
```

# Categorical Plots

```python
##Count plot

sns.countplot('day',data=df) #by default countplot plots graph by using x axis
```

```python
sns.countplot(y='sex',data=df) #for y axis give y='<column name>'
```

```python
##Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
df.head()
```

# Boxplot
 A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five number summary.

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x="day", y="total_bill",data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#categorise my data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# Violin plot

This helps us to see both distribution of data in terms of KDE and box plot

```python
sns.violinplot(data=df,x="total_bill",y="day",palette='rainbow')
```
