# CSV

```python
from io import StringIO , BytesIO 
```

```python
data = ('col1,col2,col3\n'
           'x,y,1\n'
           'a,b,2\n'
           'c,d,3')
```

```python
type(data)
```

```python
print(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
##Read from specific columns

df=pd.read_csv(StringIO(data),usecols=lambda x: x.upper() in ['COL1' , 'COL3'])
```

```python
df
```

```python
df.to_csv('abc.csv')
```

```python
##Specifying column datatypes

data=('a,b,c,d\n'
         '1,2,3,4\n'
         '5,6,7,8\n'
         '9,10,11,12')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype=object)
```

```python
df
```

```python
df['a'][1]
```

```python
df=pd.read_csv(StringIO(data),dtype=int)
```

```python
df
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':np.float,'a':'Int64'})
```

```python
df
```

```python
data_two = ('index,a,b,c\n'
               '4,apple,bat,5.7\n'
               '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data_two),index_col=0)
```

```python
data_three = ('a,b,c\n'
               '4,apple,bat,\n'
               '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data_three))
```

```python
pd.read_csv(StringIO(data_three),index_col=False)
```

```python
##Combining usecols and index_col

pd.read_csv(StringIO(data_three),usecols=['b','c'],index_col=False)
```

```python
##Quoting and Escape characters. Very useful in NLP

data_four='a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data_four),escapechar='\\')
```
